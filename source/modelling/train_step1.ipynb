# Scrapping CSV

## Imports

```python
import os
import pandas as pd
import cv2
from matplotlib import pyplot as plt
```

## Dataset

```python
images_dir_path = '../../data/dataset/'

images_paths_list = [os.path.join(path, name) for path, subdirs, files in os.walk(images_dir_path) for name in files]

images_df = pd.DataFrame(images_paths_list, columns=['path'])
```

```python
#images_df['label'] = (images_df['path'].str.extract(r'(\w+(?=/IMG))'))
images_df[['big', 'small']] = 0
images_df.loc[images_df['path'].str.contains('big'), ['big']] = 1
images_df.loc[images_df['path'].str.contains('small'), ['small']] = 1
images_df
```

## Splitting dataset

```python
train_df = images_df.loc[images_df['path'].str.contains('train')]
test_valid_df = images_df.loc[images_df['path'].str.contains('test')]

big_test_valid_df = test_valid_df.loc[test_valid_df['big']==1]
small_test_valid_df = test_valid_df.loc[test_valid_df['small']==1]

test_df = big_test_valid_df.iloc[:int(big_test_valid_df.shape[0]/2)].append(small_test_valid_df.iloc[:int(small_test_valid_df.shape[0]/2)])
valid_df = big_test_valid_df.iloc[int(big_test_valid_df.shape[0]/2):].append(small_test_valid_df.iloc[int(small_test_valid_df.shape[0]/2):])
'Train dataset:'
train_df
'Test dataset:'
test_df
'Valid dataset:'
valid_df
```

## Creating files

```python

train_df.to_csv(os.path.join(images_dir_path, 'train.csv'), index=False)
test_df.to_csv(os.path.join(images_dir_path, 'test.csv'), index=False)
valid_df.to_csv(os.path.join(images_dir_path, 'valid.csv'), index=False)
```
